```python
KORF_FILE = r"pykorf/trail_files/Cooling Water Circuit-EES-IT-LT-00141.kdf"
```

```python
from pykorf import Model, Query
from pykorf.definitions import Element, Pipe, Feed, Pump, Valve, Vessel
from pykorf.visualization import Visualizer
```

```python
model = Model(KORF_FILE)
q = Query(model)
```

```python
model.validate()
```

```python
q.by_type(Element.FEED).all()
```

```python
vis = Visualizer(model)
vis.to_html("trail.html")
```
